```python
df_long = pd.read_csv('aa_long.csv')
```

```python
df_long.head()
```

```python
df_temp = df_long.sort_values('Passengers', ascending=False)
df_temp.head(20)
```

```python
print(df_temp['Passengers'].min())
print(df_temp['Passengers'].max())
```

```python
# Smallest AA aircraft has max 50 seat -- ERJ 145
# Will filter out rows with less than 25 pax
df_temp = df_temp[df_temp['Passengers'] >= 25]
```

```python
# Largest AA aircraft has max 304 seat -- B77w
# Will filter out rows with more than 304 pax
df_temp = df_temp[df_temp['Passengers'] <= 304]
```

```python
print(df_temp['Passengers'].min())
print(df_temp['Passengers'].max())
```

```python
df_temp.columns
```

```python
df_dest = df_temp.groupby(["Origin", "Dest"])['Passengers'].sum().sort_values(ascending=False)
```

```python
df_dest.to_csv('grouped.csv')
```
