```python
# Optimized dtypes
market_dtypes = {
    'ItinID': 'int64',
    'MktID': 'int64',
    'MktCoupons': 'int8',
    'Year': 'int16',
    'Quarter': 'int8',
    'OriginAirportID': 'int32',
    'OriginAirportSeqID': 'int32',
    'OriginCityMarketID': 'int32',
    'Origin': 'str',
    'OriginCountry': 'str',
    'OriginStateFips': 'int8',
    'OriginState': 'str',
    'OriginStateName': 'str',
    'OriginWac': 'int8',
    'DestAirportID': 'int32',
    'DestAirportSeqID': 'int32',
    'DestCityMarketID': 'int32',
    'Dest': 'str',
    'DestCountry': 'str',
    'DestStateFips': 'int8',
    'DestState': 'str',
    'DestStateName': 'str',
    'DestWac': 'int8',
    'AirportGroup': 'str',
    'WacGroup': 'str',
    'TkCarrierChange': 'int8',
    'TkCarrierGroup': 'str',
    'OpCarrierChange': 'int8',
    'OpCarrierGroup': 'str',
    'RPCarrier': 'str',
    'TkCarrier': 'str',
    'OpCarrier': 'str',
    'BulkFare': 'float32',
    'Passengers': 'int32',
    'MktFare': 'float32',
    'MktDistance': 'int64',
    'MktDistanceGroup': 'int8',
    'MktMilesFlown': 'int64',
    'NonStopMiles': 'int64',
    'ItinGeoType': 'int8',
    'MktGeoType': 'int8',
}

# Define filters
focus_carriers = ['UA']

print("Loading DB1B Market data with chunk filtering...")

# Read and filter in chunks
chunk_size = 100_000
filtered_chunks = []
total_rows_read = 0
total_rows_kept = 0

for chunk in pd.read_csv(
    '../Data/DB1BMarket_2025_2.csv',
    dtype=market_dtypes,
    usecols=list(market_dtypes.keys()),
    chunksize=chunk_size,
    low_memory=False
):
    # Filter chunk
    filtered_chunk = chunk[
        (chunk['Passengers'] > 0) &
        (chunk['MktFare'] > 0) &
        (chunk['BulkFare'] == 0) &
        (chunk['OpCarrier'].isin(focus_carriers))
    ]
    
    # Track progress
    total_rows_read += len(chunk)
    total_rows_kept += len(filtered_chunk)
    
    if len(filtered_chunk) > 0:
        filtered_chunks.append(filtered_chunk)
    
    print(f"  Chunk: {len(chunk):,} rows â†’ {len(filtered_chunk):,} kept "
          f"(Total: {total_rows_read:,} read, {total_rows_kept:,} kept)")

# Combine
print("\nCombining filtered chunks...")
market_df_big8 = pd.concat(filtered_chunks, ignore_index=True)

# Memory report
memory_gb = market_df_big8.memory_usage(deep=True).sum() / 1_073_741_824
filter_rate = (total_rows_kept / total_rows_read * 100) if total_rows_read > 0 else 0

print(f"\nâœ… Loaded successfully!")
print(f"   Rows read: {total_rows_read:,}")
print(f"   Rows kept: {len(market_df_big8):,} ({filter_rate:.1f}%)")
print(f"   Columns: {len(market_df_big8.columns)}")
print(f"   Memory: {memory_gb:.2f} GB")
print(f"   Carriers: {', '.join(focus_carriers)}")

# Preview
print("\nðŸ“Š Sample data:")
print(market_df_big8[['Origin', 'Dest', 'OpCarrier', 'Passengers', 'MktFare']].head())
```

**The goal is to get the top *x* city-pairs operated by United, and solely by United. This means we are not looking at how many stops the passengers had b/w the O/D. We just care they started at the Origin and reached the Destination.**

1. Now I'll create a new column with O+D combined regardless of order
2. I'll use that to gather summary statistics for passengers (sum)
3. Then I'll create some kind of bar chart for the top 10 or 20 city-pairs.

```python
df_ua_filtered.columns
```

```python
# Create bidirectional city-pairs

df_ua_filtered['City-Pair'] = df_ua_filtered.apply(lambda x: '-'.join(sorted([x['Origin'], x['Dest']])), axis=1)

# Aggregate by City-Pair
top_markets = df_ua_filtered.groupby('City-Pair').agg({
    'Passengers': 'sum',
    'MktFare': 'mean',
    'MktDistance': 'mean',
    'MktMilesFlown': 'mean',
    'NonStopMiles': 'mean'
}).reset_index()

top_markets = top_markets.sort_values('Passengers', ascending=False)
top_markets.head()
```

```python
# Apply 10% sampling factor
SAMPLING_FACTOR = 10
top_markets['Passengers_Actual'] = top_markets['Passengers'] * SAMPLING_FACTOR

#----
# Sort and get top x
x = 20
top_markets = top_markets.sort_values('Passengers_Actual', ascending=False).head(x)

# Visualize
fig, ax = plt.subplots(figsize=(14, 6))
bars = ax.barh(top_markets['City-Pair'], top_markets['Passengers_Actual'], color='steelblue')
ax.set_xlabel('Total Passengers (Q2 2025, Estimated)', fontsize=12)
ax.set_title('Top United Airlines ' + str(x) + ' U.S. Domestic Markets by Passenger Volume\n(Based on 10% DB1B Sample)', 
             fontsize=14, fontweight='bold')
ax.invert_yaxis()

# Add value labels
for bar in bars:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height()/2, f'{width:,.0f}', 
            ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()

# Print results
print("Top 20 Markets (Q2 2025):")
print(top_markets[['City-Pair', 'Passengers', 'Passengers_Actual', 'MktFare', 'MktDistance', 'NonStopMiles']].to_string(index=False))
```
